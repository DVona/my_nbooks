```python
import seaborn as sns
import pandas as pd
```

```python
avocado = pd.read_csv('avocado.csv')
avocado.head()
```

```python
avocado.info()
```

```python
avocado.Date = pd.to_datetime(avocado.Date)
avocado.info()
```

```python
albany_avocado = avocado.query('year == 2015 and type == "conventional" and region == "Albany"')\
[['Date','Total Bags','Small Bags']]
albany_avocado.head()
```

```python
avocado_melted = pd.melt(avocado,id_vars='Date',
                                value_vars=['Total Bags','Small Bags'],
                                value_name='Count',var_name='Bags')
avocado_melted
```

```python
sns.relplot(avocado_melted,kind='line',x='Date',hue='Bags',y='Count')
```

```python
avocado['Sales Volume'] = pd.qcut(avocado['Total Volume'],q=4,labels=['Poor','Modest','Good','Excellent'])
avocado.head()
```

```python
sns.countplot(avocado,hue='Sales Volume',x='year')
```

```python
avocado_very_melted = pd.melt(avocado,id_vars='year',
                                value_vars=['Small Bags','Large Bags','XLarge Bags'],
                                value_name='Count',var_name='Bags')
avocado_very_melted
```

```python
sns.relplot(avocado_very_melted,kind='line',x='year',y='Count',hue='Bags')
```
